 # **C) Opis używanej w ćwiczeniu sieci neuronowej**

Sieci neuronowe to rodzaj algorytmu uczenia maszynowego / głębokiego uczenia się, który naśladuje sposób działania ludzkiego mózgu. Składają się z wielu warstw jednostek przetwarzających (neuronów), które łączą się ze sobą i oddziałują ze sobą poprzez funkcję aktywacji. Sieci neuronowe wykorzystują warstwy wejściowe (np. obrazy, dźwięk lub wideo), warstwy ukryte (które zawierają sztuczne neurony przetwarzające dane) i warstwy wyjściowe (które generują wynik na podstawie przetworzonych danych).

Modele sieci neuronowych są uczone przy użyciu danych wejściowych pochodzących z jednej warstwy i wyników wytwarzanych w innej warstwie, pomiędzy którymi znajduje się jedna lub więcej warstw ukrytych. Neurony pomiędzy tymi warstwami reprezentują wagi – siłę połączenia między każdym neuronem – i błędy systematyczne – wartość progową aktywacji każdego neuronu. Modele sieci neuronowych uczą się, dostosowując swoje wagi i odchylenia w czasie, aby zmniejszyć błędy w wynikach w porównaniu z oczekiwanymi wynikami. Proces ten nazywany jest szkoleniem lub propagacją wsteczną.

Sklearn *MLRegressor* to algorytm uczenia maszynowego do zadań regresyjnych. Zapewnia wysoki stopień dokładności i może obsługiwać złożone, nieliniowe zbiory danych. MLRegressor implementuje algorytm wielowarstwowego perceptronu (MLP) do uczenia i testowania zbiorów danych przy użyciu metod propagacji wstecznej i stochastycznego gradientu. Zawiera kilka parametrów, których można użyć do dostrojenia wydajności modelu, w tym liczbę warstw ukrytych, funkcje aktywacji, solwery (do optymalizacji) itp. Jest to wydajna metoda rozwiązywania problemów regresyjnych, ponieważ umożliwia uczenie się złożonych zależności nieliniowych pomiędzy zmiennymi wejściowymi i wyjściowymi.

Jedną z głównych zalet używania MLRegressor w porównaniu z innymi algorytmami regresji jest jego zdolność do obsługi danych z dużą liczbą funkcji lub danych wejściowych. Podczas gdy inne modele regresji mogą mieć problemy ze zbiorami danych zawierającymi setki lub tysiące funkcji, MLRegressor może szybko nauczyć się wzorców i zapewnić dokładne prognozy. Dzięki temu idealnie nadaje się do zastosowań takich jak przewidywanie cen akcji, wyników sprzedaży i ocen zadowolenia klientów.
<br>Kolejną zaletą korzystania z MLRegressora w porównaniu z innymi algorytmami jest jego zdolność do bezproblemowego radzenia sobie z brakującymi punktami danych, a także do automatycznego wybierania funkcji w procesie uczenia, eliminując nieistotne lub zbędne funkcje z dalszego uwzględniania w procesie budowania modelu. Ułatwia to identyfikację najważniejszych predyktorów bez konieczności ręcznego eliminowania ich ze zbioru danych, co może być dość czasochłonne w przypadku dużej liczby zmiennych wejściowych.


 GridSearchCV to proces dostrajania hiperparametrów w celu określenia optymalnych wartości dla danego modelu. Jak wspomniano powyżej, wydajność modelu w istotny sposób zależy od wartości hiperparametrów. Należy pamiętać, że nie ma możliwości poznania z góry najlepszych wartości hiperparametrów, dlatego w idealnym przypadku musimy wypróbować wszystkie możliwe wartości, aby poznać wartości optymalne. Wykonanie tego ręcznie może zająć znaczną ilość czasu i zasobów, dlatego używamy GridSearchCV do automatyzacji dostrajania hiperparametrów.

GridSearchCV to funkcja dostępna w pakiecie model_selection Scikit-learn (lub SK-learn). Ta funkcja pomaga przeglądać predefiniowane hiperparametry i dopasować estymator (model) do zestawu treningowego. Ostatecznie spośród wymienionych hiperparametrów możemy wybrać najlepsze parametry.

#Zad5

Przygotowanie do uczenia modelu.

```python
X = data.drop('csMPa', axis=1)
y = data['csMPa']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
#standaryzacja
sc = StandardScaler()
scaler = sc.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Dane zostały podzielone na część treningową (80%) i część testową (20%), a następnie ustandaryzowane na obu zbiorach za pomocą StandardScalera().

#Zad6

Próbne uczenie.

```python
regr = MLPRegressor(activation='relu',
    solver='adam',  #okresla algorytm optymalizacji wag sieci
    hidden_layer_sizes=(10, 20),    #architektura sn: liczba i rozmiar warstw ukrytych. Dwie warstwy ukryte z 10 i 20 neuronami
    alpha=0.0001,   #parametr regularyzacji, ktory pomaga zapobiegac przeuczeniu sie modelu
    random_state=20,
    early_stopping=True ,  #Ta opcja pozwala zatrzymać uczenie modelu, jeśli wyniki na zbiorze walidacyjnym przestaną się poprawiać. Pomaga to w zapobieganiu przeuczeniu.
    max_iter = 900)
regr.fit(X_train_scaled, y_train)
y_pred = regr.predict(X_test_scaled)
print('MSE: {:.2f}'.format(mean_squared_error(y_test, y_pred)))
```

Krzywa zbieżności

```python
plt.plot(regr.loss_curve_)
plt.title("Krzywa zbieżności", fontsize=14)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

Krzywa przedstawia błąd modelu na zbiorze treningowym w miarę upływu czasu(epok). Wraz ze wzrostem iteracji jego błąd spada, czyli coraz lepiej dopasowuje się do danych treningowych.

```python
errors = y_test - y_pred

plt.figure(figsize=(8, 6))
plt.hist(errors, bins=25, edgecolor='k')
plt.xlabel('Błąd')
plt.ylabel('Liczba przypadków')
plt.title('Histogram błędów')
plt.show()
```

Histogram błędów pomaga ocenić, czy reszty są rozłożone normalnie wokół zera. Jeśli histogram ma kształt zbliżony do rozkładu normalnego, to sugeruje to, że model regresji jest odpowiedni i dobrze dopasowany do danych. Natomiast jeśli histogram błędów wykazuje nietypowe wzorce, może to sugerować, że model nie radzi sobie dobrze z danymi i może wymagać korekty.
<br> Obecny histogram przypomina wykres rozkładu normalnego, wiadomo zawsze może być zawsze lepiej, ale całkiem nieźle sobie radzi.

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Wartość rzeczywista')
plt.ylabel('Wartość przewidywana')
plt.title('Wartość rzeczywista vs. Wartość przewidywana')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
```

Wykres przedstawia punkty danych na płaszczyźnie, gdzie wartości rzeczywiste są umieszczone na osi x, a wartości przewidywane przez model są umieszczone na osi y. Punkty rozproszone wokół linii y = x pokazują stopień odchylenia przewidywanych wartości od rzeczywistych wartości. Im większe jest odchylenie, tym gorsza jest jakość predykcji modelu. W najlepszym przypadku wszystkie punkty leżałyby na przerywanej linii, co oznaczałoby, że przewidywania modelu są identyczne z rzeczywistymi wartościami.
<br> Na środku tego wykresu punkty calkiem nieźle rozkładają się wzdłuż prostej, natomiast później punkty są coraz bardziej "rozproszone".

Po użyciu procedury wynik poprawił się do MSE: 34.73, więc warto jej używać.
