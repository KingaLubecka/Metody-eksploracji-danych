 # **C) Opis uywanej w wiczeniu sieci neuronowej**

Sieci neuronowe to rodzaj algorytmu uczenia maszynowego / gbokiego uczenia si, ktry naladuje sposb dziaania ludzkiego mzgu. Skadaj si z wielu warstw jednostek przetwarzajcych (neuronw), ktre cz si ze sob i oddziauj ze sob poprzez funkcj aktywacji. Sieci neuronowe wykorzystuj warstwy wejciowe (np. obrazy, dwik lub wideo), warstwy ukryte (ktre zawieraj sztuczne neurony przetwarzajce dane) i warstwy wyjciowe (ktre generuj wynik na podstawie przetworzonych danych).

Modele sieci neuronowych s uczone przy uyciu danych wejciowych pochodzcych z jednej warstwy i wynikw wytwarzanych w innej warstwie, pomidzy ktrymi znajduje si jedna lub wicej warstw ukrytych. Neurony pomidzy tymi warstwami reprezentuj wagi  si poczenia midzy kadym neuronem  i bdy systematyczne  warto progow aktywacji kadego neuronu. Modele sieci neuronowych ucz si, dostosowujc swoje wagi i odchylenia w czasie, aby zmniejszy bdy w wynikach w porwnaniu z oczekiwanymi wynikami. Proces ten nazywany jest szkoleniem lub propagacj wsteczn.

Sklearn *MLRegressor* to algorytm uczenia maszynowego do zada regresyjnych. Zapewnia wysoki stopie dokadnoci i moe obsugiwa zoone, nieliniowe zbiory danych. MLRegressor implementuje algorytm wielowarstwowego perceptronu (MLP) do uczenia i testowania zbiorw danych przy uyciu metod propagacji wstecznej i stochastycznego gradientu. Zawiera kilka parametrw, ktrych mona uy do dostrojenia wydajnoci modelu, w tym liczb warstw ukrytych, funkcje aktywacji, solwery (do optymalizacji) itp. Jest to wydajna metoda rozwizywania problemw regresyjnych, poniewa umoliwia uczenie si zoonych zalenoci nieliniowych pomidzy zmiennymi wejciowymi i wyjciowymi.

Jedn z gwnych zalet uywania MLRegressor w porwnaniu z innymi algorytmami regresji jest jego zdolno do obsugi danych z du liczb funkcji lub danych wejciowych. Podczas gdy inne modele regresji mog mie problemy ze zbiorami danych zawierajcymi setki lub tysice funkcji, MLRegressor moe szybko nauczy si wzorcw i zapewni dokadne prognozy. Dziki temu idealnie nadaje si do zastosowa takich jak przewidywanie cen akcji, wynikw sprzeday i ocen zadowolenia klientw.
<br>Kolejn zalet korzystania z MLRegressora w porwnaniu z innymi algorytmami jest jego zdolno do bezproblemowego radzenia sobie z brakujcymi punktami danych, a take do automatycznego wybierania funkcji w procesie uczenia, eliminujc nieistotne lub zbdne funkcje z dalszego uwzgldniania w procesie budowania modelu. Uatwia to identyfikacj najwaniejszych predyktorw bez koniecznoci rcznego eliminowania ich ze zbioru danych, co moe by do czasochonne w przypadku duej liczby zmiennych wejciowych.


 GridSearchCV to proces dostrajania hiperparametrw w celu okrelenia optymalnych wartoci dla danego modelu. Jak wspomniano powyej, wydajno modelu w istotny sposb zaley od wartoci hiperparametrw. Naley pamita, e nie ma moliwoci poznania z gry najlepszych wartoci hiperparametrw, dlatego w idealnym przypadku musimy wyprbowa wszystkie moliwe wartoci, aby pozna wartoci optymalne. Wykonanie tego rcznie moe zaj znaczn ilo czasu i zasobw, dlatego uywamy GridSearchCV do automatyzacji dostrajania hiperparametrw.

GridSearchCV to funkcja dostpna w pakiecie model_selection Scikit-learn (lub SK-learn). Ta funkcja pomaga przeglda predefiniowane hiperparametry i dopasowa estymator (model) do zestawu treningowego. Ostatecznie spord wymienionych hiperparametrw moemy wybra najlepsze parametry.

#Zad5

Przygotowanie do uczenia modelu.

```python
X = data.drop('csMPa', axis=1)
y = data['csMPa']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
#standaryzacja
sc = StandardScaler()
scaler = sc.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Dane zostay podzielone na cz treningow (80%) i cz testow (20%), a nastpnie ustandaryzowane na obu zbiorach za pomoc StandardScalera().

#Zad6

Prbne uczenie.

```python
regr = MLPRegressor(activation='relu',
    solver='adam',  #okresla algorytm optymalizacji wag sieci
    hidden_layer_sizes=(10, 20),    #architektura sn: liczba i rozmiar warstw ukrytych. Dwie warstwy ukryte z 10 i 20 neuronami
    alpha=0.0001,   #parametr regularyzacji, ktory pomaga zapobiegac przeuczeniu sie modelu
    random_state=20,
    early_stopping=True ,  #Ta opcja pozwala zatrzyma uczenie modelu, jeli wyniki na zbiorze walidacyjnym przestan si poprawia. Pomaga to w zapobieganiu przeuczeniu.
    max_iter = 900)
regr.fit(X_train_scaled, y_train)
y_pred = regr.predict(X_test_scaled)
print('MSE: {:.2f}'.format(mean_squared_error(y_test, y_pred)))
```

Krzywa zbienoci

```python
plt.plot(regr.loss_curve_)
plt.title("Krzywa zbienoci", fontsize=14)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

Krzywa przedstawia bd modelu na zbiorze treningowym w miar upywu czasu(epok). Wraz ze wzrostem iteracji jego bd spada, czyli coraz lepiej dopasowuje si do danych treningowych.

```python
errors = y_test - y_pred

plt.figure(figsize=(8, 6))
plt.hist(errors, bins=25, edgecolor='k')
plt.xlabel('Bd')
plt.ylabel('Liczba przypadkw')
plt.title('Histogram bdw')
plt.show()
```

Histogram bdw pomaga oceni, czy reszty s rozoone normalnie wok zera. Jeli histogram ma ksztat zbliony do rozkadu normalnego, to sugeruje to, e model regresji jest odpowiedni i dobrze dopasowany do danych. Natomiast jeli histogram bdw wykazuje nietypowe wzorce, moe to sugerowa, e model nie radzi sobie dobrze z danymi i moe wymaga korekty.
<br> Obecny histogram przypomina wykres rozkadu normalnego, wiadomo zawsze moe by zawsze lepiej, ale cakiem niele sobie radzi.

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Warto rzeczywista')
plt.ylabel('Warto przewidywana')
plt.title('Warto rzeczywista vs. Warto przewidywana')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
```

Wykres przedstawia punkty danych na paszczynie, gdzie wartoci rzeczywiste s umieszczone na osi x, a wartoci przewidywane przez model s umieszczone na osi y. Punkty rozproszone wok linii y = x pokazuj stopie odchylenia przewidywanych wartoci od rzeczywistych wartoci. Im wiksze jest odchylenie, tym gorsza jest jako predykcji modelu. W najlepszym przypadku wszystkie punkty leayby na przerywanej linii, co oznaczaoby, e przewidywania modelu s identyczne z rzeczywistymi wartociami.
<br> Na rodku tego wykresu punkty calkiem niele rozkadaj si wzdu prostej, natomiast pniej punkty s coraz bardziej "rozproszone".
